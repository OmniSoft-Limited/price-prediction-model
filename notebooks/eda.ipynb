```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Load data
data_path = "./temp/dataset.pkl"
data = pd.read_pickle(data_path)
```

```python
data.tail()
```

```python
Y = data["Price"] 
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
sns.scatterplot(x="numUsers", y="Price", data=data)
plt.title("NumUsers vs Price")
plt.show()
```

```python
categorical_features = ["softwareType", "industryDomain", "targetMarket", "adminDashboard", 
                        "uiUxDesign", "performance", "security", "availability"]

for col in categorical_features:
    plt.figure(figsize=(12,5))
    sns.boxplot(x=col, y="Price", data=data)
    plt.xticks(rotation=45)
    plt.title(f"{col} vs Price")
    plt.show()
```

```python
for col in categorical_features:
    plt.figure(figsize=(10,4))
    sns.countplot(x=col, data=data)
    plt.xticks(rotation=45)
    plt.title(f"Count of {col}")
    plt.show()
```

```python
numerical_features = ["numUsers", "timeline_months", "Price"]
plt.figure(figsize=(8,6))
sns.heatmap(data[numerical_features].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation between numerical features")
plt.show()
```
