```python
fixedAuth = []
for item in data['authentication']:
    if pd.isna(item):  
        fixedAuth.append("Null")
    elif "Multi-factor authentication" in str(item):
        fixedAuth.append("Multi-Factor")
    else:
        fixedAuth.append(item)

fixedAuth
```

```python
fixedAuthDF = pd.DataFrame(fixedAuth)
data['authentication'] = fixedAuthDF
data.head()
```

```python
data['authentication'] = (
    data['authentication']
    .fillna("None")  # replace NaN with "Null"
    .astype(str)     # ensure string for replace
    .str.replace(r"(?i).*multi-factor.*", "Multi-Factor", regex=True)  # case-insensitive replace
)
```
